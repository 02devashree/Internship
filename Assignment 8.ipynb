1. Scrape the details of most viewed videos on YouTube from Wikipedia. Url
= https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos You need to find following details: A)
Rank
B) Name
C) Artist
D) Upload date
E) Views 

```python
!pip install selenium
```

```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By

import time
```

```python
#Activating the chrome browser
driver=webdriver.Chrome()
```

```python
#Opening the home page
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos ")
```

```python
#Creating empty list for different attributes
accquried_rank=[]
video_name=[]
artist_title=[]
upload_date=[]
views_total=[]
```

```python
#Scrapping data for different attributes
rank_tags=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[1]')
for i in rank_tags:
    rank=i.text
    accquried_rank.append(rank)
    
name_tags=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[2]')
for i in name_tags:
    name=i.text
    video_name.append(name)
    
title_tags=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[3]')
for i in title_tags:
    title=i.text
    artist_title.append(title)
    
date_tags=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[5]')
for i in date_tags:
    date=i.text
    upload_date.append(date)
    
total_tags=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[4]')
for i in total_tags:
    total=i.text
    views_total.append(total)
```

```python
#Checking length for all attributes
print(len(accquried_rank),len(video_name),len(artist_title),len(upload_date),len(views_total))
```

```python
#Creating Dataframe
df=pd.DataFrame({'Rank':accquried_rank,'Name':video_name,'Title':artist_title,'Date':upload_date,'Total':views_total})
df
```

```python
driver.close()
```

8. Details of Datasets from UCI machine learning repositories.
Url = https://archive.ics.uci.edu/ You
have to find the following details:
A) Dataset name
B) Data type
C) Task
D) Attribute type
E) No of instances
F) No of attribute G) Year
 Note: - from the home page you have to go to the Show All Dataset page through code. 

```python
!pip install selenium
```

```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By

import time
```

```python
#Activating the chrome browser
driver=webdriver.Chrome()
```

```python
#Opening the home page
driver.get(" https://archive.ics.uci.edu/")
```

```python
#Clicking on Search Button-
search=driver.find_element(By.XPATH,"/html/body/div/div[1]/div[1]/main/div/div[1]/div/div/div/a[1]")
search.click()
```

```python
#Click on expand all button-
search=driver.find_element(By.CLASS_NAME,"swap")
search.click()
```

```python
#Creating empty list for different attributes:-
dataset_name=[]
data_typee=[]
given_task=[]
attribute_title=[]
numberOf_instances=[]
total_attributes=[]
mentioned_year=[]
```

```python
#scraping data for different attributes from different pages:-
start=0
end=65
for page in range(start,end):
    name=driver.find_elements(By.XPATH,'//div[@class="relative col-span-8 sm:col-span-7"]/h2/a')
    for i in name[0:10]:
        dataset_name.append(i.text)
    typee=driver.find_elements(By.XPATH,'//div[@class="grid grid-cols-8 overflow-x-auto"]/table/tbody/tr/td[2]')
    for i in typee[0:10]:
        data_typee.append(i.text)
    task=driver.find_elements(By.XPATH,'//div[@class="relative col-span-8 sm:col-span-7"]/div/div[1]')
    for i in task[0:10]:
        given_task.append(i.text)
    title=driver.find_elements(By.XPATH,'//div[@class="grid grid-cols-8 overflow-x-auto"]/table/tbody/tr/td[1]')
    for i in title[0:10]:
        attribute_title.append(i.text)
    instances=driver.find_elements(By.XPATH,'//div[@class="relative col-span-8 sm:col-span-7"]/div/div[3]/span')
    for i in instances[0:10]:
        numberOf_instances.append(i.text)
    attributes=driver.find_elements(By.XPATH,'//div[@class="relative col-span-8 sm:col-span-7"]/div/div[4]/span')
    for i in attributes[0:10]:
        total_attributes.append(i.text)
    year=driver.find_elements(By.XPATH,'//div[@class="grid grid-cols-8 overflow-x-auto"]/table/tbody/tr/td[3]')
    for i in year[0:10]:
        mentioned_year.append(i.text)
    next_button=driver.find_elements(By.XPATH,'//div[@class="btn-group"]/button[2]')
```

```python
#checking length for all attributes-
print(len(dataset_name),len(data_typee),len(given_task),len(attribute_title),len(numberOf_instances),len(total_attributes),len(mentioned_year))
```

```python
# Creating Dataframe-
df=pd.DataFrame({'Name':dataset_name,'Typee':data_typee,'Task':given_task,'Title':attribute_title,'Instances':numberOf_instances,'Attributes':total_attributes,'Year':mentioned_year})
df
```

```python
driver.close()
```
